```python
import overpy
```

```python
api = overpy.Overpass()
```

```python
result = api.query("""[out:json];
nwr["leisure"!="swimming_pool"]["access"="private"](42.633959,-1.798096,44.789632,3.365479);
out center;
""")
```

```python
coords  = []
coords += [(float(node.lon), float(node.lat)) 
           for node in result.nodes]
coords += [(float(way.center_lon), float(way.center_lat)) 
           for way in result.ways]
coords += [(float(rel.center_lon), float(rel.center_lat)) 
           for rel in result.relations]
```

```python
import requests
```

```python
for i,coord in enumerate(coords):
    if i <14000:
        lat = coord[1]
        lon = coord[0]
        r = requests.get("https://api.mapbox.com/styles/v1/mapbox/satellite-v9/static/{},{},19,0,0/512x512?access_token=pk.eyJ1IjoibWVkaGEtaCIsImEiOiJjbDNhbnlmNXAwM2liM2twOTZibXJkbms3In0.dZuVzK1P2UYS7m7aJ1aZRA".format(lon,lat))
        with open('non_pool_images/{}_{}.png'.format(lat,lon), 'wb') as output:
            output.write(r.content)
```
